#Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

#Change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-oct","5-10")
df.head()
```

gender distribution

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#from the above chart we hve analysed that:
#the number of females in the data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb,annot = True)
plt.title("Relationship b/w Parent's Education & Student's Score")
plt.show()
```

#from the above chart we have concluded that the education of the parents have a good impact on their scores

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb1,annot = True)
plt.title("Relationship b/w Parent's Marital Status & Student's Score")
plt.show()
```

#from the above chart we have concluded that there is no/negligible impact on the student's score due to their parents marital status
